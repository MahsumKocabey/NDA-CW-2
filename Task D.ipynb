```python
diagram = """
          Activity: Marathon Planning
                       |
                       v
Agent: Race Organizer ------> Entity: Marathon Route
                       |
          Activity: Running the Race
                       |
                       v
      Agent: Participant ------> Entity: Race Results
                       |
         Activity: Post-race Cleanup
"""

print(diagram)
```

```python
from prov.model import ProvDocument

# Create a new provenance document
d = ProvDocument()

# Add namespaces
d.add_namespace('prov', 'http://www.w3.org/ns/prov#')
d.add_namespace('ex', 'http://example.org/')

# Add Agents
race_organizer = d.agent('ex:RaceOrganizer')
participant = d.agent('ex:Participant')

# Add Entities
marathon_route = d.entity('ex:MarathonRoute')
race_results = d.entity('ex:RaceResults')

# Add Activities
marathon_planning = d.activity('ex:MarathonPlanning')
running_race = d.activity('ex:RunningRace')
post_race_cleanup = d.activity('ex:PostRaceCleanup')

# Add relations
d.wasGeneratedBy(marathon_route, marathon_planning)
d.wasGeneratedBy(race_results, running_race)
d.wasAssociatedWith(marathon_planning, race_organizer)
d.wasAssociatedWith(running_race, participant)
d.used(marathon_planning, marathon_route)
d.used(running_race, race_results)
d.used(post_race_cleanup, race_results)
d.wasAssociatedWith(post_race_cleanup, participant)
```

```python
# visualize the graph
from prov.dot import prov_to_dot
dot = prov_to_dot(d)
dot.write_png('marathon-prov.png')
```
